```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split

# Define Target and Feature Variables
target_column = 'Target'
y_raw = data[target_column]
X_raw = data.drop(columns=[target_column])

# Encode Categorical Variables
categorical_columns = X_raw.select_dtypes(include=['object']).columns
ohe = OneHotEncoder(drop='first', sparse_output=False)
X_categorical_encoded = ohe.fit_transform(X_raw[categorical_columns])
X_categorical_encoded = pd.DataFrame(X_categorical_encoded, columns=ohe.get_feature_names_out(categorical_columns))

# Combine Encoded Categorical and Numeric Columns
X_numeric = X_raw.select_dtypes(include=['float64', 'int64'])
X_combined = pd.concat([X_numeric.reset_index(drop=True), X_categorical_encoded.reset_index(drop=True)], axis=1)

# Scale Features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_combined)

# Apply Dimensionality Reduction
pca = PCA(n_components=0.95)  
X_preprocessed = pca.fit_transform(X_scaled)

# Encode target variable if necessary
if y_raw.dtype == 'object':
    y_preprocessed = pd.factorize(y_raw)[0]  
else:
    y_preprocessed = y_raw.values 

# Split Data into Training and Testing Sets
X_train, X_test, y_train, y_test = train_test_split(
    X_preprocessed, y_preprocessed, test_size=0.2, random_state=42
)

# Confirm the shapes
print("Training set shape:", X_train.shape)
print("Testing set shape:", X_test.shape)
```

## 2.3 Model Comparison

To compare the models based on accuracy, precision, recall, computational efficiency, and training time, we can measure each modelâ€™s training time and evaluate their performance metrics

```python
import time
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Initialize models
rf_model = RandomForestClassifier(random_state=42)
svm_model = SVC(random_state=42)
mlp_model = MLPClassifier(random_state=42, max_iter=300)

# Dictionary to store results
results = {}

# Function to train, evaluate, and record time
def evaluate_model(model, model_name):
    # Track training start time
    start_time = time.time()
    
    # Train the model
    model.fit(X_train, y_train)
    
    # Calculate training time
    training_time = time.time() - start_time
    
    # Make predictions
    y_pred = model.predict(X_test)
    
    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    
    # Store results
    results[model_name] = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1,
        'Training Time (s)': training_time
    }

# Evaluate Random Forest
evaluate_model(rf_model, 'Random Forest')

# Evaluate SVM
evaluate_model(svm_model, 'SVM')

# Evaluate MLP
evaluate_model(mlp_model, 'MLP')

# Display the results
for model_name, metrics in results.items():
    print(f"\n{model_name} Performance:")
    for metric_name, metric_value in metrics.items():
        print(f"{metric_name}: {metric_value:.4f}")
```

```python
from sklearn.model_selection import GridSearchCV
import numpy as np

# Dictionary to store results for different parameter combinations
param_results = {}

# Function to perform GridSearch and evaluate model performance
def grid_search_evaluate(model, model_name, param_grid, X_train, y_train, X_test, y_test):
    # Initialize GridSearch with cross-validation
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=1)
    grid_search.fit(X_train, y_train)
    
    # Retrieve the best model and its performance
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    training_time = grid_search.refit_time_
    
    # Record results for analysis
    param_results[(model_name, str(grid_search.best_params_))] = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1,
        'Training Time (s)': training_time
    }
    print(f"\nBest parameters for {model_name}: {grid_search.best_params_}")
    print(f"Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1 Score: {f1:.4f}")

# Parameter grids for each model
rf_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

svm_param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}

mlp_param_grid = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50)],
    'activation': ['relu', 'tanh'],
    'learning_rate': ['constant', 'adaptive'],
    'max_iter': [200, 300]
}

# Random Forest with Grid Search
rf_model = RandomForestClassifier(random_state=42)
grid_search_evaluate(rf_model, 'Random Forest', rf_param_grid, X_train_preprocessed, y_train_preprocessed, X_test_preprocessed, y_test_preprocessed)

# SVM with Grid Search
svm_model = SVC(random_state=42)
grid_search_evaluate(svm_model, 'SVM', svm_param_grid, X_train_preprocessed, y_train_preprocessed, X_test_preprocessed, y_test_preprocessed)

# MLP with Grid Search
mlp_model = MLPClassifier(random_state=42)
grid_search_evaluate(mlp_model, 'MLP', mlp_param_grid, X_train_preprocessed, y_train_preprocessed, X_test_preprocessed, y_test_preprocessed)

# Display results for parameter combinations
print("\nParameter Tuning Results:")
for (model_name, params), metrics in param_results.items():
    print(f"\n{model_name} with {params}:")
    for metric_name, metric_value in metrics.items():
        print(f"{metric_name}: {metric_value:.4f}")
```
